```python
class Student:
    def __init__(self, student_id, name, sex, weight, height):



        self.student_id = student_id
        self.name = name
        self.sex = sex
        self.weight = weight
        self.height = height
        self.courses = {}

    def enroll_course(self, course_code, course_name):

        if course_code not in self.courses:
            self.courses[course_code] = {
                "course_name": course_name,
                "midterm": {},
                "final": {}
            }
            print(f"{self.name} enrolled in {course_name} ({course_code})")
        else:
            print(f"{self.name} is already enrolled in {course_code}")

    def record_grade(self, course_code, exam_type, category, grade):

        if course_code in self.courses and exam_type in ["midterm", "final"]:
            self.courses[course_code][exam_type][category] = grade
            print(f"{exam_type.capitalize()} grade - {category}: {grade}")
        else:
            print(f"Error: Course {course_code} not found or invalid exam type.")

    def calculate_grade(self, course_code, exam_type):

        if course_code in self.courses and exam_type in ["midterm", "final"]:
            grades = self.courses[course_code][exam_type]
            weights = {"quiz": 0.20, "seatwork": 0.20, "assignment": 0.30, "recitation": 0.30}
            total_grade = 0
            total_weight = 0
            for category, weight in weights.items():
                if category in grades:
                    total_grade += grades[category] * weight
                    total_weight += weight
            if total_weight > 0:
                return total_grade


    def view_course_grades(self, course_code):

        if course_code in self.courses:
            midterm_grade = self.calculate_grade(course_code, "midterm")
            final_grade = self.calculate_grade(course_code, "final")

            print(f" {self.name}'s Grades for {self.courses[course_code]['course_name']} ({course_code}): ")
            if midterm_grade is not None:
                print(f"Midterm Grade: {midterm_grade:.2f}")
            else:
                print("Midterm Grade: Not available")

            if final_grade is not None:
                print(f"Final Grade: {final_grade:.2f}")

            else:
                print("Final Grade: Not available")


    def __str__(self):

        return (f"Student ID: {self.student_id}\n"
                f"Name: {self.name}\n"
                f"Sex: {self.sex}\n"
                f"Weight: {self.weight} kg\n"
                f"Height: {self.height} m")



if __name__ == "__main__":

    ana = Student(student_id=20230001, name="Ana", sex="Female", weight=55.0, height=1.65)
    mench = Student(student_id=20230002, name="Mench", sex="Female", weight=60.0, height=1.70)
    john = Student(student_id=20230003, name="John", sex="Male", weight=70.0, height=1.75)


    ana.enroll_course("CPE103", "Object-Oriented Programming")
    print(ana)
    ana.record_grade("CPE103", "midterm", "quiz", 85.0)
    ana.record_grade("CPE103", "midterm", "seatwork", 90.0)
    ana.record_grade("CPE103", "midterm", "assignment", 92.0)
    ana.record_grade("CPE103", "midterm", "recitation", 88.0)

    ana.record_grade("CPE103", "final", "quiz", 92.0)
    ana.record_grade("CPE103", "final", "seatwork", 88.0)
    ana.record_grade("CPE103", "final", "assignment", 95.0)
    ana.record_grade("CPE103", "final", "recitation", 90.0)
    ana.view_course_grades("CPE103")

    mench.enroll_course("CPE103", "Object-Oriented Programming")
    print(mench)
    mench.record_grade("CPE103", "midterm", "quiz", 78.0)
    mench.record_grade("CPE103", "midterm", "seatwork", 82.0)
    mench.record_grade("CPE103", "midterm", "assignment", 85.0)
    mench.record_grade("CPE103", "midterm", "recitation", 80.0)


    mench.record_grade("CPE103", "final", "quiz", 85.0)
    mench.record_grade("CPE103", "final", "seatwork", 80.0)
    mench.record_grade("CPE103", "final", "assignment", 90.0)
    mench.record_grade("CPE103", "final", "recitation", 82.0)
    mench.view_course_grades("CPE103")

    john.enroll_course("CPE103", "Object- Oriented Programming")
    print(john)
    john.record_grade("CPE103", "midterm", "quiz", 90.0)
    john.record_grade("CPE103", "midterm", "seatwork", 88.0)
    john.record_grade("CPE103", "midterm", "assignment", 95.0)
    john.record_grade("CPE103", "midterm", "recitation", 92.0)

    john.record_grade("CPE103", "final", "quiz", 82.0)
    john.record_grade("CPE103", "final", "seatwork", 90.0)
    john.record_grade("CPE103", "final", "assignment", 88.0)
    john.record_grade("CPE103", "final", "recitation", 95.0)
    john.view_course_grades("CPE103")
```
