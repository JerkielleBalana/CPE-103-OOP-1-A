```python
def convert_to_uccs_grade(final_grade):
    if final_grade >= 97:
        return 1.00
    elif final_grade >= 94:
        return 1.25
    elif final_grade >= 91:
        return 1.50
    elif final_grade >= 88:
        return 1.75
    elif final_grade >= 85:
        return 2.00
    elif final_grade >= 82:
        return 2.25
    elif final_grade >= 79:
        return 2.50
    elif final_grade >= 76:
        return 2.75
    elif final_grade >= 60:
        return 3.00
    elif final_grade >= 50:
        return 4.00
    elif final_grade >= 40:
        return 5.00
    else:
        return "Failed"


def calculate_class_standing(hands_on, quiz, assignment):
    return hands_on * 0.5 + quiz * 0.3 + assignment * 0.2


def calculate_period_grade(exam, class_standing, prev_grade_weight=0):
    period_grade = exam * 0.5 + class_standing * 0.5
    if prev_grade_weight > 0:
        return (prev_grade_weight / 3) + (period_grade * (2 / 3))
    return period_grade



STUDENTS = int(input("Enter the number of students: "))

for i in range(STUDENTS):
    print(f"\n Student {i + 1} ")
    NAME = input("Enter the name of the student: ")

    # Prelim Grades
    print("\nPRELIM GRADES")
    PHOA = float(input("Enter Prelim Hands-on Activity grade: "))
    PQUIZ = float(input("Enter Prelim Quiz grade: "))
    PAS = float(input("Enter Prelim Assignment grade: "))
    PCS = calculate_class_standing(PHOA, PQUIZ, PAS)
    print(f"Prelim Class Standing (CS): {PCS:.2f}")
    PEX = float(input("Enter Prelim Exam grade: "))
    PG = calculate_period_grade(PEX, PCS)
    print(f"Prelim Grade: {PG:.2f}")

    # Midterm Grades
    print("\nMIDTERM GRADES")
    MHOA = float(input("Enter Midterm Hands-on Activity grade: "))
    MQUIZ = float(input("Enter Midterm Quiz grade: "))
    MAS = float(input("Enter Midterm Assignment grade: "))
    MCS = calculate_class_standing(MHOA, MQUIZ, MAS)
    print(f"Midterm Class Standing (CS): {MCS:.2f}")
    MEX = float(input("Enter Midterm Exam grade: "))
    MG = calculate_period_grade(MEX, MCS, PG)
    print(f"Midterm Grade: {MG:.2f}")

    # Final Grades
    print("\nFINAL GRADES")
    FHOA = float(input("Enter Final Hands-on Activity grade: "))
    FQUIZ = float(input("Enter Final Quiz grade: "))
    FAS = float(input("Enter Final Assignment grade: "))
    FCS = calculate_class_standing(FHOA, FQUIZ, FAS)
    print(f"Final Class Standing (CS): {FCS:.2f}")
    FEX = float(input("Enter Final Exam grade: "))
    FG = calculate_period_grade(FEX, FCS, MG)
    print(f"Final Grade: {FG:.2f}")

    # Convert Final Grade to UCCS Grade
    UCCS_GRADE = convert_to_uccs_grade(FG)
    print(f"\nStudent: {NAME}")
    print(f"Prelim Grade: {PG:.2f}")
    print(f"Midterm Grade: {MG:.2f}")
    print(f"Final Grade: {FG:.2f}")
    print(f"UCCS Numerical Grade: {UCCS_GRADE}")
    print("-" * 30)
```
